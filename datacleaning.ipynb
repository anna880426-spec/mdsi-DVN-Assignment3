```python
import pandas as pd
import os


folder_path = '/content/drive/MyDrive/UTS_2026_Autumn_DVN/Assignment3'


file_name_1 = 'busperformance_reports_feb26.xlsx'
file_name_2 = 'all_modes.csv'


full_path_1 = os.path.join(folder_path, file_name_1)
full_path_2 = os.path.join(folder_path, file_name_2)

df_bus = pd.read_excel(full_path_1)
df_models = pd.read_csv(full_path_2)
```

## clean missing value

```python
df_bus_new['Region'].value_counts()
```

```python
df_bus_new['% of services cancelled'].groupby(df_bus_new['Region']).mean()
```

```python
df_bus_new[df_bus_new['Region'] == 'GS'	]
```

```python
df_bus_new['Untracked trips'] = df_bus_new['Untracked trips'].fillna(
    df_bus_new.groupby('Region')['Untracked trips'].transform('mean')
)
```

```python
df_bus_new['% of services cancelled'] = df_bus_new['% of services cancelled'].fillna(
    df_bus_new.groupby('Region')['% of services cancelled'].transform('mean')
)
```

```python
df_bus_new['Complaints per 100K'] = df_bus_new['Complaints per 100K'].fillna(
    df_bus_new.groupby('Region')['Complaints per 100K'].transform('mean')
)
```

```python
df_bus_new['Driver Vacancies'] = df_bus_new['Driver Vacancies'].fillna(
    df_bus_new.groupby('Region')['Driver Vacancies'].transform('mean')
)
```

```python
df_bus_new['OTR'] = df_bus_new['OTR'].fillna(
    df_bus_new.groupby('Region')['OTR'].transform('mean')
)
```

```python
df_bus_new
```

```python
columns_to_drop = ['Most recent month', 'Mode','FY']
df_bus_new = df_bus_new.drop(columns=columns_to_drop)

df_bus_new.head()
```

```python
df_bus_new.info()
```
